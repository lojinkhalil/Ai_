```python
# V2 - تحسين جودة البيانات باستخدام PCA

from sklearn.decomposition import PCA

# تقليل الأبعاد إلى مكونين رئيسيين
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

print("نسبة التباين :")
print(pca.explained_variance_ratio_)
```

```python
# تطبيق K-Means بعد PCA

kmeans_pca = KMeans(n_clusters=3, random_state=42)
clusters_pca = kmeans_pca.fit_predict(X_pca)
```

```python
from sklearn.metrics import silhouette_score

silhouette_v2 = silhouette_score(X_pca, clusters_pca)
inertia_v2 = kmeans_pca.inertia_

print("V2 - Silhouette Score:", round(silhouette_v2, 3))
print("V2 - Inertia:", round(inertia_v2, 2))
```

```python
# تمثيل بصري للتجميع بعد التحسين

plt.figure(figsize=(6,4))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=clusters_pca)
plt.title("K-Means Clustering after PCA (V2)")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()
```
